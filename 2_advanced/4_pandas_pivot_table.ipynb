```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

dataset = load_dataset("lukebarousse/data_jobs")
df = dataset["train"].to_pandas()

df["job_posted_date"] = pd.to_datetime(df["job_posted_date"])
```

```python
df.pivot_table(values="salary_year_avg", index="job_country", columns="job_title_short", aggfunc="median").loc[: ,["Data Analyst", "Data Engineer", "Data Scientist"]]
```

```python
df_job_country_salary = df.pivot_table(values="salary_year_avg", index="job_country", columns="job_title_short", aggfunc="median")
df_job_country_salary
```

```python
top_countries = df["job_country"].value_counts().head(6).index
top_countries
job_titles = ["Data Analyst", "Data Engineer", "Data Scientist"]
```

```python
df_job_country_salary = df_job_country_salary.loc[top_countries, job_titles]
df_job_country_salary
```

```python
df_job_country_salary.plot(kind="bar")
plt.ylabel("Salary")
plt.title("Salary per Job")
```
